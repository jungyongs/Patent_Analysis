```python
#라이브러리 불러오기
import pandas as pd

from kiwipiepy import Kiwi
```

```python
#데이터 불러오기
df=pd.read_excel('특허.xlsx',index_col=0)
df
```

## 명사 추출 하기

```python
#kiwi 객체 선언
kiwi = Kiwi()
```

```python
#일반/고유 명사 추출 함수 선언
def extract_noun(year):
    df_nouns=pd.DataFrame()
    nouns=[]
    for data in df[df['year']==year]['요약']:
        tokens=[]
        result = kiwi.tokenize(data)
        for token in result:
            if token.tag in ['NNG', 'NNP']:
                tokens.append(token.form)
        nouns.append(tokens)
        
    df_nouns['nouns']=nouns
    return df_nouns
```

## 2015년 특허 명사 추출

```python
#2015년 특허 명사 추출
nouns_2015=extract_noun(2015)
nouns_2015
```

```python
#연도별 특허 수 비교
print('2015년 특허 수: ' + str(len(df[df['year']==2015])))
print('2015년 특허 명사 개수: ' + str(len(nouns_2015)))
```

## 불용어 제거

```python
# 불용어 리스트 (특허에 형식적으로 쓰이는 표현 제거)
stopwords = ['발명','상기','포함','단계','방법','제공','설치','구비','가능','이용','복수','구성','부재',
            '특징','사용','생성','사이','이상','처리','발생','상태','개시','실시','조성','구조','단계 상기',
            '단계 포함','선택','설정','수행']
```

```python
# 불용어 제거
nouns_removed=[]
for data in nouns['nouns']:
    result=[word.replace(" ","") for word in data if word.replace(" ","") not in stopwords]
    nouns_removed.append(result)
```

```python
len(nouns_removed)
```

```python
#데이터 프레임 명사추출 행 추가
df['nouns']=nouns_removed
```
