# 연도별 특허 데이터 
- 기간: 2015년 ~ 2022년
- 데이터: 출원번호, 출원일자, 발명의 명칭, 출원인, 요약

- 권리구분: 특허, 실용
- 행정상태: 등록

```python
#컬럼명 설정
column_headers=['출원번호','출원일자','발명의 명칭','출원인','요약']
```

## 2015년 특허 데이터

```python
df_2015=pd.read_excel('2015.xlsx',names=column_headers) #데이터 불러오기
df_2015['year']=2015 #연도 컬럼 추가
df_2015
```
